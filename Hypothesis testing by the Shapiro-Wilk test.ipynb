Создание таблиц логарифмических доходностей для данных

```python
def logarithmic_yield(x,y):
  y['AMD'] = x['AMD'].apply(np.log).diff(1)
  y['INTC'] = x['INTC'].apply(np.log).diff(1)
  y['NVDA'] = x['NVDA'].apply(np.log).diff(1)
  y['ADI'] = x['ADI'].apply(np.log).diff(1)
  y['AVGO'] = x['AVGO'].apply(np.log).diff(1)
  y['ENPH'] = x['ENPH'].apply(np.log).diff(1)
  y['IPGP'] = x['IPGP'].apply(np.log).diff(1)
  y['MCHP'] = x['MCHP'].apply(np.log).diff(1)
  y['MPWR'] = x['MPWR'].apply(np.log).diff(1)
  y['MU'] = x['MU'].apply(np.log).diff(1)
  y['MXIM'] = x['MXIM'].apply(np.log).diff(1)
  y['NXPI'] = x['NXPI'].apply(np.log).diff(1)
  y['QRVO'] = x['QRVO'].apply(np.log).diff(1)
  y['SWKS'] = x['SWKS'].apply(np.log).diff(1)
  y['TXN'] = x['TXN'].apply(np.log).diff(1)
  y['XLNX'] = x['XLNX'].apply(np.log).diff(1)
  return y
```

```python
def shapiro(x,period):
  p={'Period':period}
  pv=[]
  pv.append(round(sh(x['AMD'])[1],3))
  pv.append(round(sh(x['INTC'])[1],3))
  pv.append(round(sh(x['NVDA'])[1],3))
  pv.append(round(sh(x['ADI'])[1],3))
  pv.append(round(sh(x['AVGO'])[1],3))
  pv.append(round(sh(x['ENPH'])[1],3))
  pv.append(round(sh(x['IPGP'])[1],3))
  pv.append(round(sh(x['MCHP'])[1],3))
  pv.append(round(sh(x['MPWR'])[1],3))
  pv.append(round(sh(x['MU'])[1],3))
  pv.append(round(sh(x['MXIM'])[1],3))
  pv.append(round(sh(x['NXPI'])[1],3))
  pv.append(round(sh(x['QRVO'])[1],3))
  pv.append(round(sh(x['SWKS'])[1],3))
  pv.append(round(sh(x['TXN'])[1],3))
  pv.append(round(sh(x['XLNX'])[1],3))
  for i in range(len(tickers_list)):
    p[tickers_list[i]]=pv[i]
  return  p
```

Выбор альтернативной гипотезы и оценка мощности критерия

```python
n=100000
m=[25,100,250]
```

```python
power_laplace=[]
for j in range(3):
  shapiro_laplace=[]
  power=0
  for i in range(n):
    shapiro_laplace.append(sh(sts.laplace.rvs(loc=0,scale=1,size=m[j]))[1])
  for i in shapiro_laplace:
    if i<0.05:
      power+=1
  power_laplace.append(power/n)
print(power_laplace)      
```

```python
power_t=[]
for j in range(3):
  shapiro_t=[]
  power=0
  for i in range(n):
    shapiro_t.append(sh(sts.t.rvs(df=2,loc=0,scale=1,size=m[j]))[1])
  for i in shapiro_t:
    if i<0.05:
      power+=1
  power_t.append(power/n)
print(power_t)   
```

```python
power_chi=[]
for j in range(3):
  shapiro_chi=[]
  power=0
  for i in range(n):
    shapiro_chi.append(sh(sts.chi2.rvs(df=2,loc=0,scale=1,size=m[j]))[1])
  for i in shapiro_chi:
    if i<0.05:
      power+=1
  power_chi.append(power/n)
print(power_chi)   
```

```python
x_laplas = np.linspace(sts.laplace.ppf(0.01,loc=0,scale=1),sts.laplace.ppf(0.99,loc=0,scale=1), 250)
y_laplas=sts.laplace.pdf(x_laplas,loc=0,scale=1)

x_t = np.linspace(sts.t.ppf(0.01,df=2,loc=0,scale=1),sts.t.ppf(0.99,df=2,loc=0,scale=1), 250)
y_t = sts.t.pdf(x_t,df=2,loc=0,scale=1)

x_chi2 = np.linspace(sts.chi2.ppf(0.01,loc=0,scale=1,df=2),sts.chi2.ppf(0.99,df=2,loc=0,scale=1), 250)
y_chi2 = sts.chi2.pdf(x_chi2,df=2,loc=0,scale=1)

x_norm = np.linspace(sts.norm.ppf(0.01,loc=0,scale=1),sts.norm.ppf(0.99,loc=0,scale=1), 250)
y_norm=sts.norm.pdf(x_norm,loc=0,scale=1)

fig, ax = plt.subplots(1, 1)

ax.plot(x_laplas,y_laplas,color='red')
ax.plot(x_t,y_t, color='green')
ax.plot(x_chi2,y_chi2, color='purple')
ax.plot(x_norm,y_norm, color='blue')

ax.legend(['Laplas','t','chi2','norm'])
```
